```python
from sklearn import datasets
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import seaborn as sns
import matplotlib.pyplot as plt

# Load Iris Dataset
iris = datasets.load_iris()
df = pd.DataFrame(iris['data'], columns=iris['feature_names'])
print(df.head())
```

```python
# Standardize the Data
scaler = StandardScaler()
scaled_data = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
```

```python
# Correlation before PCA
plt.figure(figsize=(6, 4))
sns.heatmap(scaled_data.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix (Before PCA)")
plt.show()
```

```python
# PCA Transformation
pca = PCA(n_components=3)
pca.fit(scaled_data)
data_pca = pca.transform(scaled_data)

data_pca = pd.DataFrame(data_pca, columns=['PC1', 'PC2', 'PC3'])
print(data_pca.head())
```

```python
# Correlation after PCA
plt.figure(figsize=(6, 4))
sns.heatmap(data_pca.corr(), annot=True,fmt='.2f', cmap='Greens')
plt.title("Correlation Matrix (After PCA)")
plt.show()
```
